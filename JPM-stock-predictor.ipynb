```python
import yfinance as yf
```

```python
jpm= yf.Ticker("jpm")
```

```python
jpm = jpm.history(period="max")
```

```python
jpm.plot.line(y="Close", use_index=True)
```

```python
jpm["Tomorrow"] = jpm["Close"].shift(-1)
```

```python
jpm
```

```python
jpm["Target"] = (jpm["Tomorrow"] > jpm["Close"]).astype(int)
```

```python
jpm
```

```python
jpm = jpm.loc["2000-01-01":].copy()
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = jpm.iloc[:-100]
test= jpm.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

predictions = model.predict(test[predictors])
```

```python
import pandas as pd 

predictions = pd.Series(predictions, index=test.index)
```

```python
precision_score(test["Target"], predictions)
```

```python
combined = pd.concat([test["Target"], predictions], axis=1)
```

```python
combined.plot()
```

```python
def predict(train, test, predictors, model): 
        model.fit(train[predictors], train["Target"])
        predictions = model.predict(test[predictors])
        predictions = pd.Series(predictions, index=test.index, name = "Predictions")
        combined = pd.concat([test["Target"], predictions], axis=1)
        return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(jpm, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
horizons = [2, 5, 60, 250, 1000]

new_predictors = []

for horizon in horizons:
    rolling_averages = jpm.rolling(horizon).mean()
    
    ratio_column = f"Close Ratio_{horizon}"
    jpm[ratio_column] = jpm["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    jpm[trend_column] = jpm.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column, trend_column]
```

```python
jpm = jpm.dropna()
```

```python
jpm
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model): 
        model.fit(train[predictors], train["Target"])
        predictions = model.predict_proba(test[predictors])[:,1]
        predictions[predictions >= .6] = 1
        predictions[predictions < .6] = 0
        predictions = pd.Series(predictions, index=test.index, name = "Predictions")
        combined = pd.concat([test["Target"], predictions], axis=1)
        return combined
```

```python
predictions = backtest(jpm, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
